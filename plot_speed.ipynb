```python
import pandas as pd
csv_file = pd.read_csv('./wave.csv')
csv_df = pd.DataFrame(csv_file)
```

```python
csv_df
```

```python
df_12_inc = csv_df[csv_df['inc12'] == 1]
```

```python
# Incident under the 60 min
df_12_inc['real12'].plot(label='Ground Truth')
ax = df_12_inc['pred12'].plot(label='Prediction')
ax.legend()
```

```python
# Non-incident under the 60 min
df_12_non = csv_df[csv_df['inc12'] == 0]
df_12_non['real12'].plot(label='Ground Truth')
ax = df_12_non['pred12'].plot(label='Prediction')
ax.legend()
```
